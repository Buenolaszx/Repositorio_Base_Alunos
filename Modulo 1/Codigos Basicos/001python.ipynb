## O que são variáveis

Variáveis são espaços na memória do computador onde podemos guardar informações(como textos, números, resultados de cálculos etc) para usar depois

Variável com texto (string)

```python
nome = "Miguel"
print(nome)
```

```python
nome = "Bueno"
```

```python
print(nome)
```

Variável com número inteiro (int)

```python
idade = 16
print(idade)
```

para fazer um comentário em um código, coloque # para não altera-lo

```python
# este é um exemplo de um número inteiro negativo
temperatura = -3 # esta variável está armazenando um número negativo
print(temperatura) # print() é uma função de saída
```

```python
# para o melhor desempenho durante a programação, evite usar acentos, "ç" e etc
producao = "20 caixas"
print(producao)
```

Variável com número decimal (float)

```python
altura = 1.78 # em python, os números decimais são escritos com "."
# exemplo = 3.14
print(altura)
```

```python
pi = 3.14
print(pi)
```

Variável com valor booleano (bool)

```python
estudante = True
print(estudante)
```

## Nomes de variáves

Pode:

*   Começar com letra (Exemplo: N, Número, Altura, etc)
*   Pode conter um número (Exemplo: N1, Número1, etc)

Não pode:

*   Começar com número (Exemplo: 1N, 1Número)
*   Usar espaço (Exemplo: nome_completo, nome-completo, NomeCompleto)
*   Usar acentuações ou caracteres especiais (@, #, $, %, &)
*   Usar palavras reservadas (palavras que desempenham papéis fundamentais na estruturação e funcionalidade do código)

## O que é snake_case?

snake_case é um estilo de escrita usado para nome de variáveis, funções, e nomes em geral no código. Neste estilo, as palavras são todas em minúsculo e separadas por um underline ( _ )

## Por que usar snake_case?


*  Deixar o código mais fácil de ler
*  É o padrão recomendado pela comunidade Python
*  Segue a PEF8, que é o guia de estilo oficial do Python

OBS: PEF8 é um documento que recomenda boas práticas para escrever código Python limpo e legível.




## Exemplos que não usam snake_Case
