# Função print(   )
A função print(   ) é uma das funções mais importantes e usadas na linguagem Python. Sua função é, basicamente, exibir mensagens na tela ou enviá-las para outro dispositivo, como imprimir dentro de um arquivo de texto

No Python 3, print(   ) é uma função interna, de modo que não é necessário importar nenhuma bíblioteca para poder utiliza-lá. Basta chamá-la e passar os argumentos necessários.

```python
print("Meu primeiro print") # Todo texto deve estar entre aspas simples ou duplas
```

```python
print('Meu segundo print')
```

```python
print("D'água") # Quando uma palavra contém aspas, as aspas iniciais devem ser diferentes
```

```python
print(20 + 15) # O Python realiza a operação que está dentro dos parenteses
```

```python
print("20 + 15") # A operação entre aspas será apenas impressa, não executada, pois o python entende isso como um texto
```

```python
print("20 + 15 =", 20 + 15)
```

```python
n1 = "1"
n2 = "3"
print(n1 + n2) # Neste exemplo, o Python faz a concatenação (unir) das 2 strings
```

```python
nome = "Miguel"
sobrenome = "Bueno"
nome_completo = nome + " " + sobrenome
print(nome_completo)
```
