1) Crie um código que receba o nome do aluno, duas notas e informe a média do aluno.
obs: no seu print você deve apresentar o nome do aluno, as duas notas e a média.

```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
media =(nota1 + nota2)/2
nomeDoAluno = (input("Digite o nome do Aluno: "))
print(f"O(a) aluno(a) {nomeDoAluno} teve notas {nota1}, {nota2} e sua média é {media}")
```

2) Crie um código que receba as medidas de um retângulo e informe o perímetro do mesmo

```python
vertice1 = float(input("Digite a medida da altura do retângulo: "))
vertice2 = float(input("Digite a medida da largura do retângulo: "))
perimetro = (vertice1 + vertice2)*2
print(f"O perímetro do retângulo é igual a {perimetro}")
```

3) Crie um código que calcule a raiz quadrada de um número

```python
raiz = float(input("Digite o número que deseja calcular: "))
raizQuadrada = (raiz) **0.5
print(f"A raiz quadrada de {raiz} é igual a {raizQuadrada}")
```

4) Crie uma calculadora simples (adição, subtração, multiplicação e divisão). Seu código deve receber 2 números e apresentar o resultado das 4 operações

```python
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
soma = (numero1 + numero2)
subtracao = (numero1 - numero2)
multiplicacao = (numero1 * numero2)
divisao = (numero1 / numero2)
print(f"Os resultados são:soma {soma}, da subtração {subtracao}, da multiplicação {multiplicacao}, e da divisão {divisao}")
```

5) Calcule a diferença de dois números

```python
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
diferenca = (num1 - num2)
print(f"A diferença dos números é {diferenca}")
```

7) Crie um código que receba uma base e depois um expoente e calcule a potência

8) Crie um programa que solicite ao usuário a quantidade de horas trabalhadas por semana e o valor da hora do trabalho e calcule o salário mensal

```python
hora = float(input("Digite a quantidade de horas trabalhadas na semana: "))
valor = float(input("Digite o valor da hora trabalhada: "))
salario_mensal = ((hora*4)*(valor))
salario_anual = (salario_mensal *12)
print(f"O salário mensal é de R${salario_mensal:.2f} e o salário anual é R${salario_anual:.2f}")
```
