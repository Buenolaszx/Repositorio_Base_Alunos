# Operadores Relacionais


   
 * (>) Maior
 * (>=) Maior igual
 * (<) Menor
 * (<=) Menor igual
 * (!=) Diferente
 * (==) Igual








```python
10 > 20 # 10 é maior que 20
```

```python
10 < 20 # 10 é menor que 20
```

```python
5+1 != 6 # 5+1 é diferente que 6
```

```python
3+1 >= 4 # 3+1 é maior ou igual a 4
```

```python
nome_aluno = "Miguel"
nome_professor = "Alexsandra"
nome_aluno == nome_professor
```

# Operadores Lógicos



*   AND (e)
*   OR (ou)
*   NOT (negação)






# AND

true + true = true

true + false = false

false + true = false

false + false = false

```python
(10 > 20) and (20 > 30) #false + false = false
```

```python
(10 > 20) and (20 < 30) #false + true = false
```

```python
(10 < 20) and (20 > 30) #true + false = false
```

```python
(10 < 20) and (20 < 30) #true + true = true
```

# OR
True + True = True

True + False = True

False + True = True

False + False = False

```python
(1 < 2) or (3 > 6)
```

```python
"Miguel" == "Miguel" or "Python" != "python"
```
